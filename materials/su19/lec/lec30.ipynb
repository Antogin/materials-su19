### Fitted Values to Observed Y

What is the ratio of how the fitted/predicted values vary compared to the actual y values? 

$\frac{\text{SD of fitted values}}{\text{SD of observed y}} = |r|$

### SD of Residuals
$\text{SD of residuals} = \sqrt{1-r^2}\cdot\text{SD of y}$

```python
# How do residuals vary with a correlation of 1?
x = np.arange(0, 51, 1)
y = x
t = Table().with_columns("x", x, "y", y)
t.scatter("x", fit_line=True)
```

```python
# How do residuals vary with a correlation of 0? 
x = np.random.normal(50, 5, 100)
xlims = np.array([np.min(x), np.max(x)])
errors = np.random.normal(0, 6, 100)
y = (0 * x + 10) + errors
sample = Table().with_columns('x', x, 'y', y)
sample.scatter("x")
plots.plot(xlims, 0*xlims + 10, lw=4);
```
